```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats  # For statistical tests
# %matplotlib inline
```

```python

import os

# 1. Load cleaned datasets
countries = ['benin', 'sierraleone', 'togo']
dfs = {}

for country in countries:
    file_path = f'data/processed/{country}_clean.csv'
    if os.path.exists(file_path):
        dfs[country] = pd.read_csv(file_path, parse_dates=['Timestamp'], index_col='Timestamp')
    else:
        print(f"Warning: File not found for {country}: {file_path}")
```

```python
#  2. Metric Comparison - Boxplots
metrics = ['GHI', 'DNI', 'DHI']
plt.figure(figsize=(15, 10))

for i, metric in enumerate(metrics, 1):
    plt.subplot(1, 3, i)
    data = pd.concat([dfs['benin'][metric], 
                     dfs['sierraleone'][metric], 
                     dfs['togo'][metric]], 
                    axis=1, 
                    keys=countries)
    sns.boxplot(data=data)
    plt.title(f'{metric} Distribution by Country')
    plt.ylabel('W/m')
    plt.xlabel('Country')
    
plt.tight_layout()
plt.show()
```
